```python
def initialize_client():
    return replicate.Client(api_token="Your_replicate_token_here")
```

```python
def fetch_patient_data(patient_id, data):
    patient_data = data[data["Patient_ID"] == patient_id]
    if patient_data.empty:
        return None
    return patient_data
```

```python
def create_single_param_prompt(patient_id, param_name, raw_summary):
    prompt = f"""
Analyze the data provided and give a direct and professional summary of trends, abnormalities, and actionable recommendations. Ensure the output:
1. Includes exactly 1 actionable recommendations, numbered clearly.
2. Use short sentences of length 15 to 20 words, ans use simple words
3. The summary should be easy to read and understand.


**Patient Details:**
- Patient ID: {patient_id}

**{param_name} Raw Summary:**
{raw_summary}

"""
    return prompt
```

```python
def generate_insight_for_param(patient_id, param_name, raw_summary, client):
    prompt = create_single_param_prompt(patient_id, param_name, raw_summary)
    print(f"Processing {param_name}...")
    try:
        output = client.run(
            "meta/llama-2-7b-chat",
            input={"prompt": prompt, "max_length": 512, "temperature": 0.5}
        )
        return " ".join(output).strip()
    except Exception as e:
        print(f"An error occurred while processing {param_name}: {e}")
        return f"Unable to process {param_name}."
```

```python
def generate_combined_summary(patient_id, patient_data, client):
    global raw_summary_global, generated_summary_global

    gender = "Male" if patient_data["Gender"].iloc[0] == 1 else "Female"
    age = round(patient_data["Age"].iloc[0])
    max_iculos = patient_data["ICULOS"].max()


    summary_header = f"Patient ID: {patient_id}\nGender: {gender}\nAge: {age} years\nICULOS: {max_iculos} hours\n\n"

    # Analyze each parameter using existing analysis functions
    analyses = {
        "Heart Rate (HR)": analyze_heart_rate(patient_data),
        "Oxygen Saturation (O2Sat)": analyze_o2saturation(patient_data),
        "Body Temperature (Temp)": analyze_body_temperature(patient_data),
        "Systolic Blood Pressure (SBP)": analyze_systolic_blood_pressure(patient_data),
        "Mean Arterial Pressure (MAP)": analyze_mean_arterial_pressure(patient_data),
        "Respiration Rate (RR)": analyze_respiration_rate(patient_data)
    }


    raw_summary_global = {param_name: analysis_result for param_name, analysis_result in analyses.items()}

    insights = []
    for param_name, raw_summary in raw_summary_global.items():
        insight = generate_insight_for_param(patient_id, param_name, raw_summary, client)
        insights.append(f"**{param_name}**\n{insight}")

    generated_summary_global = summary_header + "\n\n".join(insights)

    return raw_summary_global, generated_summary_global
```

```python
def main():
    global raw_summary_global, generated_summary_global
    global patient_id

    # Ask for Patient ID
    patient_id = int(input("Enter the Patient ID: "))
    data = pd.read_csv('/content/drive/MyDrive/Cse 570/Df_Medical_Anomalies.csv')

    column_mapping = {
        "Heart Rate": "HR",
        "Oxygen Saturation": "O2Sat",
        "Body Temperature": "Temp",
        "Systolic Blood Pressure": "SBP",
        "MAP": "MAP",
        "Respiratory Rate": "Resp"
    }

    data.rename(columns=column_mapping, inplace=True)

    patient_data = fetch_patient_data(patient_id, data)
    if patient_data is None:
        print(f"No data found for Patient ID {patient_id}.")
        return

    client = initialize_client()

    raw_summary_global, generated_summary_global = generate_combined_summary(patient_id, patient_data, client)

    # Print the raw and generated summaries
    print("\nRaw Summaries:")
    for param, summary in raw_summary_global.items():
        print(f"{param}:\n{summary}\n")

    print("\nGenerated Medical Summary:")
    print(generated_summary_global)


if __name__ == "__main__":
    main()
```

```python
def clean_output(summary):

    lines = summary.splitlines()

    cleaned_lines = []
    for line in lines:

        line = " ".join(line.strip().split())
        if line.startswith("*"):
            line = line.replace("-", "").strip()
        if line:
            cleaned_lines.append(line)

    cleaned_summary = "\n".join(cleaned_lines)

    return cleaned_summary
cleaned_output = clean_output(generated_summary_global)

```

```python
import textwrap

def format_summary(summary, width=100):
    lines = summary.splitlines()
    formatted_lines = [textwrap.fill(line, width) for line in lines]
    return "\n".join(formatted_lines)

formatted_summary = format_summary(cleaned_output)
```

```python
from fpdf import FPDF
import matplotlib.pyplot as plt
import os

class PDF(FPDF):
    def __init__(self):
        super().__init__()
        self.set_auto_page_break(auto=True, margin=15)

    def add_summary(self, summary):
        self.add_page()
        self.set_font("Arial", size=12)
        for line in summary.splitlines():
            self.multi_cell(0, 10, line)
            self.ln(2)

    def add_plot(self, image_path, title):
        self.add_page()
        self.set_font("Arial", size=14)
        self.cell(0, 10, title, ln=True, align="C")
        self.ln(10)
        self.image(image_path, x=10, y=30, w=180)

def save_plots_to_images(data, patient_id, parameter_columns, hour_column="Hour", output_folder="plots"):

    os.makedirs(output_folder, exist_ok=True)


    patient_data = data[data["Patient_ID"] == patient_id]
    if patient_data.empty:
        print(f"No data found for Patient ID {patient_id}.")
        return []


    if hour_column not in patient_data.columns:
        print(f"The hour column '{hour_column}' is not present in the dataset.")
        return []

    image_paths = []
    for param in parameter_columns:
        if param not in patient_data.columns:
            print(f"The parameter '{param}' is not present in the dataset.")
            continue

        plt.figure(figsize=(10, 6))
        plt.plot(patient_data[hour_column], patient_data[param], marker='o', linestyle='-', label=param)
        plt.xlabel("Hour")
        plt.ylabel(param)
        plt.title(f"Hour vs {param} for Patient ID {patient_id}")
        plt.legend()
        plt.grid(True)

        image_path = os.path.join(output_folder, f"{param}_Patient_{patient_id}.png")
        plt.savefig(image_path)
        plt.close()
        image_paths.append(image_path)

    return image_paths

def save_summary_and_plots_to_pdf(summary, plot_images, output_pdf_path):
    """
    Save the generated summary and plots into a single PDF.

    Parameters:
        summary (str): The generated summary to include in the PDF.
        plot_images (list): List of file paths for the plot images.
        output_pdf_path (str): The path to save the output PDF.
    """
    pdf = PDF()
    pdf.add_summary(summary)

    for image_path in plot_images:
        title = os.path.basename(image_path).replace("_", " ").replace(".png", "").title()
        pdf.add_plot(image_path, title)

    pdf.output(output_pdf_path)
    print(f"Generated summary and plots saved to {output_pdf_path}")


def main():

    data = pd.read_csv('/content/drive/MyDrive/Cse 570/Df_Medical_Anomalies.csv')
    column_mapping = {
        "Heart Rate": "HR",
        "Oxygen Saturation": "O2Sat",
        "Body Temperature": "Temp",
        "Systolic Blood Pressure": "SBP",
        "MAP": "MAP",
        "Respiratory Rate": "Resp"
    }
    data.rename(columns=column_mapping, inplace=True)


    parameter_columns = ["HR", "O2Sat", "Temp", "SBP", "MAP", "Resp"]

    plot_images = save_plots_to_images(data, patient_id, parameter_columns)


    summary =  formatted_summary

    output_pdf_path = "Generated_Summary_Patient.pdf"
    save_summary_and_plots_to_pdf(summary, plot_images, output_pdf_path)


if __name__ == "__main__":
    main()
```

```python
import textstat
from detoxify import Detoxify


# Function to calculate Gunning Fog
def calculate_readability_metrics(text):
    return {
        "Gunning Fog Index": textstat.gunning_fog(text)
    }


# Function to calculate toxicity
def analyze_toxicity(text):
    results = Detoxify("original").predict(text)
    return results

readability_metrics = calculate_readability_metrics(formatted_summary)
toxicity_analysis = analyze_toxicity(formatted_summary)

print("Readability Metrics:", readability_metrics)
print("Toxicity Analysis:", toxicity_analysis)
```

```python
from transformers import pipeline, AutoTokenizer
import math

model_name = "cross-encoder/nli-deberta-base"
classifier = pipeline("text-classification", model=model_name, device=-1)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def split_text_to_chunks(text, max_tokens=512):
    """
    Tokenize and split text into chunks of max_tokens length.
    """
    tokens = tokenizer.tokenize(text)
    chunks = []
    for i in range(0, len(tokens), max_tokens - 50):
        chunk = tokens[i:i + max_tokens - 50]
        chunks.append(tokenizer.convert_tokens_to_string(chunk))
    return chunks

def rank_summary(summary, metric_name, max_tokens=512):
    """
    Evaluate the summary by splitting it into smaller chunks and averaging scores.
    """
    chunks = split_text_to_chunks(summary, max_tokens)
    scores = []
    for i, chunk in enumerate(chunks):
        prompt = f"""
        You are tasked with evaluating a medical summary based on its {metric_name}.
        Score the summary between 0 (poor) and 1 (excellent) for {metric_name}.
        Summary: {chunk}
        Your evaluation:
        """
        try:
            result = classifier(prompt, truncation=True)
            score = float(result[0]['score'])
            print(f"Chunk {i+1}/{len(chunks)} - Score: {score}")
            scores.append(score)
        except Exception as e:
            print(f"Error processing chunk {i+1}: {e}")
            continue


    final_score = sum(scores) / len(scores) if scores else 0.0
    return final_score

if __name__ == "__main__":


    completeness_score = rank_summary(formatted_summary, "completeness")
    coherence_score = rank_summary(formatted_summary, "coherence")

    print(f"Completeness Score: {completeness_score}")
    print(f"Coherence Score: {coherence_score}")
```
